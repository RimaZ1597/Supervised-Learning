```python
import pandas as pd

data = {
    'Student': [1, 2, 3, 4, 5, 6, 7, 8],
    'Studied': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No'],
    'Passed': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
df
```

```python
#==================================================
P_Pass = len(df[df['Passed'] == 'Yes']) / len(df)
#==================================================
P_studey_given_Pass= len(df[(df['Studied'] == 'Yes') & (df['Passed'] == 'Yes')]) / len(df[df['Passed'] == 'Yes'])
P_studey_given_Pass
#==================================================
# Calculate  probabilities
P_studey_given_Pass = len(df[(df['Studied'] == 'Yes') & (df['Passed'] == 'Yes')]) / len(df[df['Passed'] == 'Yes'])
P_Not_studey_given_Pass = len(df[(df['Studied'] == 'No') & (df['Passed'] == 'Yes')]) / len(df[df['Passed'] == 'Yes'])
P_Studied_given_Not_Passed = len(df[(df['Studied'] == 'Yes') & (df['Passed'] == 'No')]) / len(df[df['Passed'] == 'No'])
P_Not_Studied_given_Not_Passed = len(df[(df['Studied'] == 'No') & (df['Passed'] == 'No')]) / len(df[df['Passed'] == 'No'])
P_Not_Studied_given_Pass = len(df[(df['Studied'] == 'No') & (df['Passed'] == 'Yes')]) / len(df[df['Passed'] == 'Yes'])

#========================================================================
P_Studied = len(df[df['Studied'] == 'Yes']) / len(df)
P_Not_Studied = len(df[df['Studied'] == 'No']) / len(df)
#========================================================================
P_Passed_given_Studied = (P_Not_Studied_given_Pass * P_Pass) / P_Studied
P_Passed_given_Not_Studied = (P_Not_Studied_given_Not_Passed * P_Pass) / P_Not_Studied
#========================================================================
print(f"P(Passed) = {P_Pass}")
print(f"P(Studied | Passed) = {P_studey_given_Pass}")
print(f"P(Not Studied | Passed) = {P_Not_studey_given_Pass}")
print(f"P(Studied | Not Passed) = {P_Studied_given_Not_Passed}")
print(f"P(Not Studied | Not Passed) = {P_Not_Studied_given_Not_Passed}")
print(f"P(Studied) = {P_Studied}")
print(f"P(Not Studied) = {P_Not_Studied}")
print(f"P(Passed | Studied) = {P_Passed_given_Studied}")
print(f"P(Passed | Not Studied) = {P_Passed_given_Not_Studied}")
print(f"P(Studied | Passed) = {P_Passed_given_Not_Studied}")
#========================================================================
```
